La página web que vamos a scrappear será: https://books.toscrape.com/

```python
import requests
from bs4 import BeautifulSoup

url_base = 'https://books.toscrape.com/catalogue/page-'
current_page = 1  # Nuestra base de partida es la página 1
proceed = True
data = []      #Creamos una lista donde guardaremos los datos que vamos scrappeando.

while proceed:
    print(f"Procesando página {current_page}")
    url = url_base + str(current_page) + ".html"

    page = requests.get(url)

    if page.status_code == 404:  # Verifica si la página no existe
        proceed = False
    else:
        soup = BeautifulSoup(page.text, 'html.parser')

        # Encontrar todos los libros en la página
        all_books = soup.find_all("li", class_="col-xs-6 col-sm-4 col-md-3 col-lg-3")
        for book in all_books:
            item = {}
            # Extraemos el título
            item["title"] = book.find("img")["alt"]
            # Extraemos el link
            item["link"] = book.find("a")["href"]
            # Extraemos el precio
            item["price"] = book.find("p", class_="price_color").text[2:]
            # Extraemos la disponibilidad
            item["stock"] = book.find("p", class_="instock availability").text.strip()

            # Agregamos cada libro a la lista de datos creada al principio
            data.append(item)

        current_page += 1

# Mostramos los datos extraidos
for book in data:
    print(book)
```
