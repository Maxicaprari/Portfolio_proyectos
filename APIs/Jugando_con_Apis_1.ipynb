```python
from nba_api.stats.static import teams
import matplotlib.pyplot as plt
```

```python
def one_dict(list_dict):
    keys=list_dict[0].keys()
    out_dict={key:[] for key in keys}
    for dict_ in list_dict:
        for key, value in dict_.items():
            out_dict[key].append(value)
    return out_dict
```

Este código define una función llamada one_dict que toma una lista de diccionarios (list_dict) y los combina en un solo diccionario.
Lo que hace específicamente:

Obtiene las claves del primer diccionario de la lista con list_dict[0].keys()
Crea un nuevo diccionario (out_dict) donde:

Cada clave es una clave del primer diccionario
Cada valor es una lista vacía


Recorre cada diccionario en la lista original
Para cada par clave-valor en cada diccionario, añade el valor a la lista correspondiente en out_dict
Devuelve el diccionario combinado

```python
nba_teams = teams.get_teams()

nba_teams[0:3]
```

```python
import pandas as pd
dict_nba_team=one_dict(nba_teams)
df_teams=pd.DataFrame(dict_nba_team)
df_teams.head()
```

```python
df_warriors=df_teams[df_teams['nickname']=='Warriors']
df_warriors
```
